## Handling The Outliers 

```python
for colname,colvals in bc.iteritems():
    if bc[colname].dtype!=object and colname!='diagnosis':
        Q1=bc[colname].quantile(q=0.25)
        Q3=bc[colname].quantile(q=0.75)
        IQR=Q3-Q1
        low=Q1-(1.5*IQR)
        up=Q3+(1.5*IQR)
        bc.loc[(bc[colname]<=low) | (bc[colname]>=up) ,colname]
        print(colname)
    else:
        continue
```

```python
for colname,colvals in bc.iteritems():
    if bc[colname].dtype!=object and colname!='diagnosis':
        Q1=bc[colname].quantile(q=0.25)
        Q3=bc[colname].quantile(q=0.75)
        IQR=Q3-Q1
        low=Q1-(1.5*IQR)
        up=Q3+(1.5*IQR)
        bc.loc[(bc[colname]<low) | (bc[colname]>up) ,colname]=bc[colname].median()
    else:
        continue
```

## Handling the categorical data 

```python
bc=pd.get_dummies(bc,drop_first=True)
```

```python
bc.head()
```

## Machine Learning Algorithm 

```python
# select x & y
# x=independent columns
# y=Dependent columns(Traget columns)
X=bc.drop('diagnosis_M',axis=1)
y=bc['diagnosis_M']
```

```python
# stadardization
# To comes all Data into one unit excpet Traget columns.Data comes in between(0,1)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
# split the data in traininig and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
# model selation
# import the model
# Dataset Target columns is Binary hence we go for LogisticRegression 
from sklearn.linear_model import LogisticRegression
 
# instantiate the model
logreg=LogisticRegression()

# fit the model with Training Data
logreg.fit(x_train,y_train)

# prediction of Data
y_train_pred=logreg.predict(x_train)
y_test_pred=logreg.predict(x_test)
```

```python
from sklearn import metrics
print("Training Accuracy:",metrics.accuracy_score(y_train,y_train_pred))
print("Testing Accuracy:",metrics.accuracy_score(y_test,y_test_pred))
```

```python
# confusion metrix
# Training Data
from sklearn.metrics import confusion_matrix
cnf_metrix_train=confusion_matrix(y_train,y_train_pred)
cnf_metrix_train
```

```python
from sklearn import metrics
print("Training Precision:",metrics.precision_score(y_train,y_train_pred))
print("Testing Precision:",metrics.precision_score(y_test,y_test_pred))
print("Training Recall:",metrics.recall_score(y_train,y_train_pred))
print("Testing Recall:",metrics.recall_score(y_test,y_test_pred))
```

```python
Training_result=metrics.classification_report(y_train,y_train_pred)
print(Training_result)
```

```python
Testing_result=metrics.classification_report(y_test,y_test_pred)
print(Testing_result)
```
